```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
data = pd.read_csv('data/bank-additional-full-preprocessed.csv')
```

__Visualizing Age Distribution__

```python
sns.histplot(x="age", data=data , kde=True, hue= "y")
plt.title("Age Distribution and Deposits\n")
plt.show()
```

__Outlier Detection__

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize each numerical column using boxplots
numerical_columns = ['age', 'campaign', 'pdays', 'previous',
                     'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

for column in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data[column])
    plt.title(f"Boxplot for {column}")
    plt.show()
```

__Visualizing correlation matrix__

```python
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

cols = data.select_dtypes("object").columns

le = LabelEncoder()

data[cols] = data[cols].apply(le.fit_transform)

plt.figure(figsize=(23,10))
sns.heatmap(data.corr(), cmap='bwr', annot=True)
plt.show()
```
